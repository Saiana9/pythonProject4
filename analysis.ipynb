```python
#Импортируем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Загружаем данные по зарплате за период с 2000 по 2016 г
data2000_2016 = pd.read_csv('/Users/sayana/Documents/Data_for_final_project/2000-2016 гг.-Tаблица 1.csv', sep=';')
data2000_2016
```

```python
#Формируем отдельно данные по выбранным видам деятельности за период с 2000 по 2016 г
years = range(2000, 2017, 1)
строительство = data2000_2016.iloc[24, 1:].values
образование = data2000_2016.iloc[32, 1:].values
здравоохранение = data2000_2016.iloc[33, 1:].values
data_dict2000_2016 = {}
data_dict2000_2016['Год'] = years
data_dict2000_2016['Строительство номинальная зп'] = строительство
data_dict2000_2016['Образование номинальная зп'] = образование
data_dict2000_2016['Здравоохранение номинальная зп'] = здравоохранение
data2000_2016 = pd.DataFrame(data_dict2000_2016)
data2000_2016
```

```python
#Загружаем данные по зарплате за период с 2017 по 2023 г
data2017_2023 = pd.read_csv('/Users/sayana/Documents/Data_for_final_project/с 2017 г.-Tаблица 1.csv', sep=';')
data2017_2023                
```

```python
#Формируем данные по выбранным видам деятельности за период с 2017 по 2023 г
years = range(2017, 2024, 1)
строительство = data2017_2023.iloc[39, 1:].values
образование = data2017_2023.iloc[50, 1:].values
здравоохранение = data2017_2023.iloc[51, 1:].values
data_dict2017_2023 = {}
data_dict2017_2023['Год'] = years
data_dict2017_2023['Строительство номинальная зп'] = строительство
data_dict2017_2023['Образование номинальная зп'] = образование
data_dict2017_2023['Здравоохранение номинальная зп'] = здравоохранение
data2017_2023 = pd.DataFrame(data_dict2017_2023)
data2017_2023
```

```python
#Соединяем данные за весь период
df = pd.concat([data2000_2016, data2017_2023], join="inner")
df.index = range(0,24,1)
df
```

```python
#Построим графики изменения зарплаты по годам для этих видов экономической деятельности
df.plot(x="Год", y=["Строительство номинальная зп", "Образование номинальная зп", "Здравоохранение номинальная зп"], figsize=(30,20))
plt.title('Изменение номинальной зарплаты по годам', fontsize=50)
plt.legend()
plt.grid()
plt.show()
```

```python
#Выводы: За период с 2000 по 2018 годы наибольшая номинальная заработная плата составляла заработная плата в строительстве.
#Далее с 2018 по 2021 годы заработная плата в сфере здравоохранения превысила заработную плату в сфере строительства. Наименьшая номинальная заработная плата на протяжении всего анализируемого периода в сфере образования. 
```

```python
#Загружаем данные по инфляции
data_inflation = pd.read_csv('/Users/sayana/Documents/Data_for_final_project/Таблица уровня инфляции по месяцам в годовом исчислении.csv', sep=';')
data_inflation
```

```python
#Добавляем столбец с инфляцией в данные по зарплате
inf = data_inflation.loc[1:24, 'Всего']
df['Инфляция'] = inf.iloc[::-1].reset_index(drop=True)
df
```

```python
#Посчитаем Индекс инфляции и реальные зарплаты с учетом инфляции, сохраняем в новые стобцы
df['Индекс инфляции'] = df.apply(lambda x: (x['Инфляция']/100 +1), axis=1)
df['Строительство реальная зп'] = df.apply(lambda x: x['Строительство номинальная зп']/ x['Индекс инфляции'], axis=1)
df['Образование реальная зп'] = df.apply(lambda x: x['Образование номинальная зп']/ x['Индекс инфляции'], axis=1)
df['Здравоохранение реальная зп'] = df.apply(lambda x: x['Здравоохранение номинальная зп']/ x['Индекс инфляции'], axis=1)
df = df[['Год', 'Строительство номинальная зп', 'Строительство реальная зп', 'Образование номинальная зп', 'Образование реальная зп', 'Здравоохранение номинальная зп', 'Здравоохранение реальная зп', 'Инфляция', 'Индекс инфляции']]
cols = ['Строительство реальная зп','Образование реальная зп','Здравоохранение реальная зп']
df[cols] = df[cols].round(1)
df
```

```python
#Построим графики изменения реальной зарплаты по годам по видам деятельности
df.plot(x="Год", y=["Строительство реальная зп", "Образование реальная зп", "Здравоохранение реальная зп"], figsize=(30,20))
plt.title('Изменения реальной зарплаты по годам', fontsize=50)
plt.legend()
plt.grid()
plt.show()
```

```python
#Выводы: На графике видно, что реальная заработная плата в строительстве больше чем в сфере здравоохранения и образования. Но в период с 2018 по 2022 годы реальная заработная плата в здравоохранении превышала заработную плату в стрительстве.
```
